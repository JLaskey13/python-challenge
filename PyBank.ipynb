```python
import csv
import os

total_months = 0
total_revenue = 0
average_revenue_change = 0
revenue_increase = ["",0]
revenue_decrease = ["",0]
greatest_increase = ["",0]
greatest_decrease = ["", 999999999]
previous_revenue = ["",0]
total_revenue_change = 0

budget_data = os.path.join("","budget_data_1.csv")
file_to_output = os.path.join("", "budget_analysis_1.txt") 

with open(budget_data) as revenue_data:
    reader = csv.reader(revenue_data)
    
    # Read the header row
    header = next(reader)
    
    first_row = next(reader)
    
    #pulling second column in first record
    previous_revenue = int(first_row[1])
    total_revenue = int(first_row[1])
    
    for row in reader:
        total_months = total_months + 1
        total_revenue = total_revenue + int(row[1])
        revenue_increase = int(row[1]) - previous_revenue
        revenue_decrease = int(row[1]) - previous_revenue
        previous_revenue = int(row[1])
        current_month = row[0]
   
    # Comparing each revenue increase to current greatest increase 
    
        if revenue_increase > greatest_increase[1]:
            greatest_increase[1] = revenue_increase
            greatest_increase[0] = current_month
    
    # Comparing each revenue decrease to current greatest decrease
    
        if revenue_decrease < greatest_decrease[1]:
            greatest_decrease[1] = revenue_decrease
            greatest_decrease[0] = current_month
        
    # Add all revenue changes and divide by total months - 1, Store revenue changes in list 

        total_revenue_change = total_revenue_change + revenue_increase
        average_revenue_change = total_revenue_change / total_months
        
    output = (f"Total Months {total_months}\n"
              f"Total Revenue ${total_revenue}\n"
              f"Average Change in Revenue: ${average_revenue_change}\n"
              f"Greatest Increase in Revenue: {current_month} {greatest_increase[1]}\n "
              f"Greatest Decrease in Revenue: {current_month} {greatest_decrease[1]}")
    
    print(output)
    
    
    
    
```
