```python
import pandas as pd

df = pd.read_csv('employee_data1.csv')
df.head()

# Change the date format:

def conv_date (input_date_string):
    year, month, day = input_date_string.split('-')
    return "{}/{}/{}".format(month, day, year)
    
df["Correct DOB"] = df["DOB"].map(conv_date)

# Masked SSN:

def hide_ssn (ssn_str):
    return "***-**-" + ssn_str[-4:]

df["Hidden SSN"] = df["SSN"].map(hide_ssn)

# Split Names:

def first_name (name_string):
    return name_string.split(' ')[0]

df["First Name"] = df["Name"].map(first_name)

def last_name (name_string):
    return name_string.split(' ')[1]

df["Last Name"] = df["Name"].map(last_name)

df.head()
```
